 ###### STEP 1 : Importing libraries

```python
import pandas as pd   # importing pandas 
import numpy as np      # importing numpy
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

###### STEP 2 : Adding Dataset

```python
df = pd.read_csv("content/data_2_var.csv", header=None)   #reading csv data
df.tail()   # showing last row data
```

```python
df.shape
```

```python
import seaborn as sns  # importing seaborn  
%matplotlib inline
sns.heatmap(df.corr())     # Using heat map to select the important features
```

```python
sns.heatmap(df.isnull())  #heatmap of uncleared data i.e with null values
```

```python
df.isnull().sum()   #  now no null values
```

```python
plt.scatter(df[0], df[1])
plt.show()
```

```python
df.corr()
```

```python
X = np.array(df[0]).reshape(-1, 1)
y = np.array(df[1]).reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)

regr = LinearRegression()
 
regr.fit(X_train, y_train)
print(regr.score(X_test, y_test))
```

```python
y_pred = regr.predict(X_test)
plt.scatter(X_test, y_test, color ='b')
plt.plot(X_test, y_pred, color ='k')
 
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
degree=3
polyreg=make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyreg.fit(X_train,y_train)
print(polyreg.score(X_test, y_test))
```

```python
import matplotlib.pyplot as plt
X_seq = np.linspace(X.min(),X.max(),300).reshape(-1,1)
plt.figure()
plt.scatter(X_test, y_test)
plt.plot(X_seq,polyreg.predict(X_seq),color="black")
plt.title("Polynomial regression with degree "+str(degree))
plt.show()
```
